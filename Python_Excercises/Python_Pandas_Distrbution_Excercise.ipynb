1. The average number of applications submitted for a standard exam is 600 every 3 months, representing a Poisson distribution. Write a python program to plot the probability that 100 to 500 applications will be submitted for that exam next month. Provide an appropriate title and X and Y axis labels. (3 points) 

```python
from scipy.stats import poisson 
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
rate=600/3 
n=np.arange(100,500)
y=stats.poisson.pmf(n, rate)
plt.plot(n, y, 'o-')
plt.title('Poisson : $\lambda$ = %i' %rate)
plt.xlabel('Num of applications')
plt.ylabel('Applications Probability')
plt.show()
```

2. The probability of Mary successfully completing her university course is 40%. Write a Python program to plot a PMF of the binomial distribution for 30 such courses. Give the appropriate title, X and Y axis labels (3 points) 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
no_of_courses =30
probability =0.4
k=np.arange(0,30)
binomial=stats.binom.pmf(k,no_of_courses,probability)
plt.plot(k, binomial, 'o-')
plt.title('Binomial Distribution: n =%ip =%f' %(no_of_courses,probability))
plt.xlabel('Number of courses')
plt.ylabel('Probability of successfully completion')
plt.show()
```

3. Write a python program to plot a PDF of normally distributed data ranging from -30 to 70 with a mean of 42 and standard deviation of 6. (2 points)

```python
mean_value =42     
Stan_dev =6  
x=np.arange(-30,70,0.1)
y=stats.norm.pdf(x,mean_value, Stan_dev)
plt.plot(x,y)
plt.show()
```

4. Suppose we have a full deck of 52 cards, and 2 cards are picked without replacement. Write a Monte Carlo simulation to run this experiment 300 times. Success is the act of getting a numeric card (cards from 2 to 10) for the first pick and an alphabetic card (cards J, Q, K, or A) for the second pick. Calculate the probability of success of the entire experiment. (6 points). 

```python
import random
import numpy as np
def RandomPick():
    return np.random.choice([2,3,4,5,6,7,8,9,10,'A','K','Q','J'],1)
def CardPick(numTrials):
    numpicks =0
    for i in range(numTrials):
        c1=RandomPick()
        c2=RandomPick()
        if c1[0].isnumeric()and not c2[0].isnumeric():
            numpicks+=1
    print("Success Probaility of Experiment =", numpicks/numTrials) 
CardPick(300)
```

5. Using the tips data (tips.csvDownload tips.csv), which contains details of tips offered to a waiter at a restaurant, write a Python program that will sample 50 tips from male customers and 50 tips from female customers. Plot the histogram for each sample, side by side in a 1x2 subplot. Provide an appropriate title and X and Y axis labels for each histogram (6 points)

```python
import pandas as pd
import random
df = pd.read_csv("tips.csv")
df
Gender = df['sex']
Tip = df['tip']
g1 = Gender == 'Female'
g2 = Gender == 'Male'
t1 = Tip[g1].dropna()
t2 = Tip[g2].dropna()
t1 = random.sample(list(t1),50)
t2 = random.sample(list(t2),50)
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(1, figsize = (10,6))
plt.subplots_adjust(hspace = 0.5)

plt.subplot(1,2,1)
plt.hist(t1,color = 'yellow', edgecolor = 'grey')
plt.xlim(0,10)
plt.title("Female Tips Distribution")
plt.xlabel('Female ')
plt.ylabel('Tips ')

plt.subplot(1,2,2)
plt.hist(t2, color = 'grey', edgecolor = 'yellow')
plt.xlim(0,10)
plt.title("Male Tips Distribution")
plt.xlabel('Male ')
plt.ylabel('Tips ')
```
