2.Define a class with one method that finds the minimum element in the list, and another method that finds the maximum element in the list. Find and print the minimum and maximum elements of the list [5, 2, 1, 4, 3] by creating an object of your class and calling the respective methods. (3 points)

```python
class MinMax:
    def _init_(self, list):
        self.input_list = list

    def getMin (self, input_list):
        min_value = min(input_list)
        print('Minimum element is:', min_value)

    def getMax (self, input_list):
        max_value = max(input_list)
        print('Maximum element is:', max_value)

list = [5, 2, 1, 4, 3]
object= MinMax()
object.getMin(list)
object.getMax(list)
```
